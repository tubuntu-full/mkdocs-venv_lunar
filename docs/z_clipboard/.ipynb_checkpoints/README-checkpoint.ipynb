**Warning: When using Jupyter Notebooks**. . .      
MkDocs with plugins like mkdocs-jupyter and themes such as ReadTheDocs often use *the first heading in the Jupyter Notebook (usually the content of the first Markdown cell with an H1 or H2 heading)* as the title displayed in the navigation menu.

If a notebook has a heading, it takes precedence over the filename.

If you want to use HTML inline within a Markdown file (e.g., to embed custom components), 
you can mix Markdown and raw HTML. For example:

When you run mkdocs serve to start the MkDocs development server, localhost:8000 corresponds to the root directory of the docs/ folder within your MkDocs project, not the root of your repository or the site/ folder.

```python
my-mkdocs-project/
├── docs/                   # Content served at localhost:8000
│   ├── index.md            # Root page of the site
│   ├── video/
│   │   ├── lunar.mp4
│   │   ├── lunarorbit.webm
│   │   ├── lunar2.png
│   └── other-page.md
├── mkdocs.yml              # MkDocs configuration file
└── site/                   # Output folder generated by `mkdocs build`
```

Roboto Mono Fonts
- Add to <head> section


<!DOCTYPE html>
<html lang="en">
<head>
  <meta charset="UTF-8">
  <meta name="viewport" content="width=device-width, initial-scale=1.0">
  <title>Scoped Roboto Mono Font</title>
  <link href="https://fonts.googleapis.com/css2?family=Roboto+Mono:wght@400;700&display=swap" rel="stylesheet">
</head>
<body>
  <p>This is normal text.</p>
  <p><span style="font-family: 'Roboto Mono', monospace;">This is text in Roboto Mono.</span></p>
</body>
</html>
